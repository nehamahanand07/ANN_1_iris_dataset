```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Iris.csv')
df
```

```python
#in ann output should also be numeric, so encode. reason: activation function is performed
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Species'] = le.fit_transform(df['Species'])
df
```

```python
#dropping id
df.drop(['Id'],axis=1,inplace=True)
```

```python
#sep x and y
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#CONVERT DATA INTO TRAINING AND TESTING DATA
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
y_train
x_test
y_test

```

```python
#NORMALIZATION OR SCALING
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()  #object created
scaler.fit(x_train)     # fitting training input bcz in equation considers x-training data
x_scaler_train=scaler.transform(x_train) #transforming x_train
x_scaler_test=scaler.transform(x_test) #transforming x_test
```

```python
#model creation
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()

#input layer, units=4 as 4 features
model.add(Dense(units=4,activation='relu'))

#hidden layer

model.add(Dense(units=5,activation='relu'))

#hidden layer
model.add(Dense(units=4,activation='relu'))

#output 3 so softmax
model.add(Dense(units=3,activation='softmax'))
```

```python
#compile loss is sparse as not binary
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_scaler_train,y_train,epochs=75,batch_size=10)
```

```python
model.evaluate(x_scaler_test,y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#in sigmoid 0.5 above=1 and below is 0
#in softmax there are 3 values
y_new=model.predict(scaler.transform([[2.3,1,5,3]]))
y_new
```

```python
#using argmax to find highest value's index
ind=y_new.argmax()
ind
```

```python
#we got 2 but we need to reconvert to get species name as we encoded earlier
pred=le.inverse_transform([ind])
pred
```

```python
y_pred1=model.predict(x_scaler_test)
#print(y_pred1)
y_pred2=y_pred1.argmax(axis=1) #result of y_pred1 but highest index collected to get exact answer
y_pred2
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred2)
print(cr)
```
